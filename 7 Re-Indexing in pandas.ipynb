```python
import pandas as pd
import numpy as np
df=pd.read_excel(r"W:\JupyterWS\file_example_10.xls")
df=df.loc[(df["Gender"]=="Female") & (df["Age"]<35)]
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
df.sort_index(ascending=False,inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
index=[0,3,2,1]
reindexed_df=df.reindex(index)
reindexed_df
```

```python
col=['Last Name','First Name','Gender','Country','Date','Id']
reindexed_column=df.reindex(col,axis=1)
reindexed_column
```

```python
col=['Last Name','First Name','Gender','Country','Date','Id']
reindexed_column=df.reindex(col,axis=1,fill_value='India')
reindexed_column
```
